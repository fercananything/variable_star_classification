```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
empty = []

file = open("/Users/zoesurles/Desktop/variable_star_classification/downloaded_lightcurves/frequencies.txt")
lines = file.readlines()
try:
    for line in lines:
        line = line.split(" ")
        line[2] = line[2].replace("\n", " ").strip()
        empty.append(1/float(line[2]))
except:
    print("error")
    print(line)

print(empty)
```

```python
plt.figure(figsize=(10, 8))

plt.hist(empty, bins=70, range=None, weights=None, density=None)
plt.xlabel('Period (days)')
plt.ylabel('Number of Stars with that Period')
plt.title('Period Distribution of 1945 RR Stars')
```
